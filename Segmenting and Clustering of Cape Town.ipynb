```python
url = 'https://en.wikipedia.org/wiki/List_of_Cape_Town_suburbs'
page = urlopen(url).read().decode('utf-8')
soup = BeautifulSoup(page, 'html.parser')
my_div = soup.find('div',{'class':'mw-parser-output'})
table = soup.body.table.tbody
```

<h2> Extract the data from from wikipedia page </h2>

```python


def get_cell(element):
    cells = element.find_all('td')
    row = []
    
    for cell in cells:
        if cell.a:            
            if (cell.a.text):
                row.append(cell.a.text)
                continue
        row.append(cell.string.strip())
        
    return row

def get_row():    
    data = []  
    
    for tr in table.find_all('tr'):
        row = get_cell(tr)
        if len(row) != 3:
            continue
        data.append(row)        
    
    return data
```
